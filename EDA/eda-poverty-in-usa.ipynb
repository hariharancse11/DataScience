```python
# Imports
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data
df = pd.read_csv('/kaggle/input/poverty-level-wages-in-the-usa-dataset-1973-2022/poverty_level_wages.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
plt.plot(df['year'],df['annual_poverty-level_wage'])
plt.xlabel('Year')
plt.ylabel('Poverty level wage')
plt.title('year vs poverty level wage')
plt.show()
```

## This plot shows that in 1970's if a person's wage is 5000 or below the he is under poverty and in 2020's it stands around 25000

```python
plt.plot(df['year'],df[['white_share_below_poverty_wages','black_share_below_poverty_wages','hispanic_share_below_poverty_wages']])
plt.legend(['white_share_below_poverty_wages','black_share_below_poverty_wages','hispanic_share_below_poverty_wages'])
plt.show()
```

## The above plot shows that black and hispanic people's poverty is higher that white's throughout the period the there is a drastic reduction in poverty in 2010's

```python
sns.barplot(x=df['year'],y=df['hourly_poverty-level_wage'])
plt.xticks(rotation=90)
plt.show()
```

## Hourly wage increased from 2 to 12 dollars

```python
sns.scatterplot(x=df['year'],y=df['white_women_share_below_poverty_wages'])
sns.scatterplot(x=df['year'],y=df['white_men_share_below_poverty_wages'])
plt.legend(['white_women_share_below_poverty_wages','white_men_share_below_poverty_wages'])
plt.show()
```

## This scatter plot shows that on average white men are in less poverty level

```python
# Visualization 1: Pairplot to explore relationships between numerical columns
sns.pairplot(df[['annual_poverty-level_wage', 'hourly_poverty-level_wage',
                '0-75%_of_poverty_wages', '75-100%_of_poverty_wages', 'share_below_poverty_wages']])
plt.show()
```

```python
# Visualization 2: Boxplot for the distribution of annual poverty-level wages
sns.boxplot(data=df, y='annual_poverty-level_wage')
plt.show()


```

```python
# Visualization 3: Histogram for hourly poverty-level wages
sns.histplot(data=df, x='hourly_poverty-level_wage', bins=20, kde=True)
plt.show()

```

```python
# Visualization 4: Barplot for different poverty wage categories
poverty_wage_columns = ['0-75%_of_poverty_wages', '75-100%_of_poverty_wages', 'share_below_poverty_wages']
sns.barplot(data=df[poverty_wage_columns], ci=None)
plt.xticks(rotation=45)
plt.show()

```

```python
# Visualization 5: Distribution of 100-125% of poverty wages
sns.kdeplot(data=df, x='100-125%_of_poverty_wages', shade=True)
plt.show()
```
